```
kdd98data.drop(['AGEFLAG', 'DOB', 'GEOCODE2', 'LIFESRC', 'MAILCODE', 'NOEXCH', 'ODATEDW', 'OSOURCE', 'SOLP3', 'STATE', 'ZIP', 'TCODE'], axis=1, inplace=True)
```

```
feat_to_rm= [c for c in kdd98data.columns if c[:5] == 'RAMNT' or c[:5] == 'RDATE' ]
```

```
kdd98data.drop(feat_to_rm, axis=1, inplace=True)
```

```
#Converting to date format
kdd98data['LASTDATE']=kdd98data['LASTDATE'].astype(str) + '01'
kdd98data['ADATE_2']=kdd98data['ADATE_2'].astype(str) + '01'
kdd98data['MAXADATE']=kdd98data['MAXADATE'].astype(str)+ '01'

kdd98data['LASTDATE']=pd.to_datetime(kdd98data['LASTDATE'],format='%y%m%d',errors='coerce')
kdd98data['ADATE_2']=pd.to_datetime(kdd98data['ADATE_2'],format='%y%m%d',errors='coerce')
kdd98data['MAXADATE']=pd.to_datetime(kdd98data['MAXADATE'], format='%y%m%d', errors='coerce')
```

```
#DERIVED Features

# calculate the time difference between LAST_DATE and ADATE_2
from datetime import timedelta

kdd98data['TOT_WEEKS']=round((abs(kdd98data['LASTDATE']-kdd98data['ADATE_2']).dt.days)/7,2)
kdd98data['TOT_WEEKS']


#similarly, you can calculate the last promotion to gift time gap -- MAXADATE, ADATE_2
kdd98data['TOT_MAX_WEEKS']=round((abs(kdd98data['MAXADATE']-kdd98data['ADATE_2']).dt.days)/7,2)
kdd98data['TOT_MAX_WEEKS']


#average value of all responses, average value of response to all card promotions
kdd98data.loc[kdd98data.NUMPROM>0,'AVG_ALL_RESPONSE']=kdd98data['NGIFTALL']/kdd98data['NUMPROM']
kdd98data.loc[kdd98data.NUMPROM <= 0,'AVG_ALL_RESPONSE']=0


kdd98data.loc[kdd98data.CARDPROM>0,'AVG_CARD_RESPONSE']=kdd98data['CARDGIFT']/kdd98data['CARDPROM']
kdd98data.loc[kdd98data.CARDPROM <= 0,'AVG_CARD_RESPONSE']=0


#Similarly, you can create a new variable for last gift to max gift ratio

kdd98data.loc[kdd98data.MAXRAMNT>0,'LAST_MAX_GIFT']=kdd98data['LASTGIFT']/kdd98data['MAXRAMNT']
kdd98data.loc[kdd98data.MAXRAMNT <= 0,'LAST_MAX_GIFT']=0

#another new variable for ratio of max to min gift amount
kdd98data.loc[kdd98data.MAXRAMNT>0,'MIN_MAX_AMNT']=kdd98data['MINRAMNT']/kdd98data['MAXRAMNT']
kdd98data.loc[kdd98data.MAXRAMNT <= 0,'MIN_MAX_AMNT']=0



kdd98data[['AVG_ALL_RESPONSE', 'AVG_CARD_RESPONSE','LAST_MAX_GIFT','MIN_MAX_AMNT','TOT_WEEKS','TOT_MAX_WEEKS']]

```

```
feat_to_rm= [c for c in kdd98data.columns if c[:6] == 'ADATE_' or c[:4] == 'RFA_' ]

def Diff(li1, li2): 
    return (list(set(li1) - set(li2))) 
RFA=['RFA_2R','RFA_2F','RFA_2A']
feat_to_rm=Diff(feat_to_rm,RFA)
```

```
kdd98data.drop(feat_to_rm,axis=1,inplace=True)
```

```
kdd98data.loc[kdd98data.HOMEOWNR == 'H', 'HOMEOWNR'] = 1
kdd98data.loc[kdd98data.HOMEOWNR != 1 , 'HOMEOWNR'] = 0
```

```
def mutate(feat):
  kdd98data.loc[kdd98data[feat]=='X', feat]= 1
  kdd98data.loc[kdd98data[feat] !=1, feat] = 0
```

```
kdd98data.drop(['CHILD03', 'CHILD07','CHILD12','CHILD18'], axis=1, inplace=True)
```

```
mutate('RECINHSE')
mutate('RECP3')
mutate('RECPGVG')
mutate('RECSWEEP')
mutate('MAJOR')
mutate('PEPSTRFL')
mutate('RECPGVG')
```

```
kdd98data['RECPGVG'].unique()
```

```
#Replacing NA values with 0 in these variables
interestVars=['VETERANS', 'BIBLE','CATLG', 'HOMEE', 'PETS', 'CDPLAY', 'STEREO','PCOWNERS', 'PHOTO', 'CRAFTS', 'FISHER', 'GARDENIN', 'BOATS', 'WALKER', 'KIDSTUFF', 'CARDS', 'PLATES']

for i in interestVars:
  kdd98data[i].fillna(0, inplace=True)
```

```
kdd98data['PLATES'].unique()
```

```
#Replacing Y with 1
def Replace_y(var):
  kdd98data.loc[kdd98data[var]=='Y', var]=1

for i in interestVars:
  Replace_y(i)
```

```
#Replacing NaN with -1
#the variables like MBBOOKS, MBCRAFT, etc give the number of known times the donor has responded to other  types of mail order offers.' -- so maybe the missing values here should be set to -1 (to differentiate from 0)

interestVars=['MBCRAFT', 'MBGARDEN', 'MBBOOKS', 'MBCOLECT', 'MAGFAML', 'MAGFEM', 'MAGMALE', 'PUBGARDN', 'PUBCULIN', 'PUBHLTH', 'PUBDOITY', 'PUBNEWFN', 'PUBPHOTO', 'PUBOPP']

for i in interestVars:
  kdd98data[i].fillna(-1, inplace=True)
```

```
#Replace strings with 'missing'
interestVars=['CLUSTER', 'GENDER', 'CLUSTER2', 'DOMAIN',  'WEALTH1', 'WEALTH2']

for i in interestVars:
  kdd98data[i].fillna('missing', inplace=True)
```

```
#the DOMAIN variable has 2 characters, first one for Urbanicity and the secoond for Socio-economic status. We want to separate these out into two variables

kdd98data['DOMAINU'] = kdd98data['DOMAIN'].str.get(0)
kdd98data['DOMAINSES'] = kdd98data['DOMAIN'].str.get(1)
```

```
#Replacing i with '5' in DOMAINSES so that it can be converted to int 

kdd98data.loc[kdd98data.DOMAINSES=='i', 'DOMAINSES']='5'
```

```
# Converting DOMAINSES to integer type

kdd98data['DOMAINSES']=kdd98data['DOMAINSES'].astype(int)
```

```
#remove some of the original attributes used for deriving new attributes
kdd98data.drop(['LASTDATE','MAXADATE','HOMEOWNR','MDMAUD_R','AGE902','HHAS2','CONTROLN','HPHONE_D','NGIFTALL','DOMAIN','DOMAINU', 'WEALTH2','WEALTH1'], axis=1, inplace=True)
```

```
TARGET_B=kdd98data['TARGET_B']
TARGET_D=kdd98data['TARGET_D']
kdd98data.drop(['TARGET_B','TARGET_D'], axis=1, inplace=True)
```

```
kdd98data_tidy['AGE']=kdd98data_tidy['AGE'].fillna(kdd98data_tidy['AGE'].median())
kdd98data_tidy['INCOME']=kdd98data_tidy['INCOME'].fillna(kdd98data_tidy['INCOME'].median())
kdd98data_tidy['DATASRCE']=kdd98data_tidy['DATASRCE'].fillna('3')
```

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(kdd98data_tidy['RFA_2A'].astype(str))
kdd98data_tidy['RFA_2A']=le.transform(kdd98data_tidy['RFA_2A'].astype(str))
kdd98data_tidy['RFA_2A']=le.transform(kdd98data_tidy['RFA_2A'].astype(str))
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(kdd98data_tidy, TARGET_B, test_size=0.3, random_state=42)
```

```
## Import the random forest Classifier.
from sklearn.ensemble import RandomForestClassifier 
rf = RandomForestClassifier(n_estimators=100) 
rf.fit(X_train, y_train)
```

```
rf.score(X_train, y_train)
```

```
rf.score(X_test, y_test)
```

```
import pandas as pd
feature_importances = pd.DataFrame(rf.feature_importances_,
                                   index = X_train.columns,
                                    columns=['importance']).sort_values('importance', ascending=False)
                                    
feature_importances.reset_index(inplace=True)
feature_importances
```

```
feature_importances['importance'][:127].sum()
```

```
Top_50_features=feature_importances['index'][0:50]
X_train=X_train[Top_50_features]
X_test=X_test[Top_50_features]
```

```
#Parameters used by our current forest

rf = RandomForestClassifier()
from pprint import pprint
print('Parameters currently in use:\n')
pprint(rf.get_params())
```

```
#Hyperparameter Tuning grid

from sklearn.model_selection import RandomizedSearchCV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 200, num = 5)]

# Number of features to consider at every split
max_features = ['auto', 'sqrt']

# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 50, num = 5)]
max_depth.append(None)

# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]

# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]

# Method of selecting samples for training each tree
bootstrap = [True, False]

# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
pprint(random_grid)
```

```
from sklearn.model_selection import RandomizedSearchCV
# Using random grid to search for best hyperparameters
# base model
rf = RandomForestClassifier()
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 5, cv = 3, verbose=2, random_state=42, n_jobs = 1)
# Fit the random search model
rf_random.fit(X_train, y_train)
```

```
#Best Parameters
rf_random.best_params_
```

```
#Optimized Random Forest
rf = RandomForestClassifier(n_estimators=125, bootstrap=True, max_depth=20, max_features='auto', min_samples_leaf=4, min_samples_split=2) 
rf.fit(X_train, y_train)
```

```
rf.score(X_train, y_train)
```

```
rf.score(X_test,y_test)
```

```
y_pred_test=rf.predict(X_test)
y_pred_train=rf.predict(X_train)
```

```
#Confusion matrix
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
cm=confusion_matrix(y_train,y_pred_train)

sns.heatmap(cm, square=True, annot=True, cbar=False, cmap='Wistia', linewidths=0.2, linecolor='k' )
plt.xlabel('predicted value')
plt.ylabel('true value');
```

```
cm=confusion_matrix(y_test,y_pred_test)

sns.heatmap(cm, square=True, annot=True, cbar=False, cmap='Wistia', linewidths=0.2, linecolor='k' )
plt.xlabel('predicted value')
plt.ylabel('true value');
```

```
#ROC plot
import sklearn.metrics as metrics

# calculate the fpr and tpr for all thresholds of the classification
probs = rf.predict_proba(X_test)
preds = probs[:,1]
fpr, tpr, threshold = metrics.roc_curve(y_test, preds)
roc_auc = metrics.auc(fpr, tpr)

# method I: plt
import matplotlib.pyplot as plt
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
